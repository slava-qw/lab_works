```python
a = df[(t_1 <= df['time']) & (df['time'] <= t_2)]
a.reset_index(drop=True, inplace=True)
a
```

```python
t_start = a['time'][0]
t_end = a['time'][len(a) - 1]
T = t_end - t_start
w = 2 * np.pi / T
w, T
```

```python
a['time_norm'] = a['time'] - t_start
```

```python
phi_0 = np.pi / 6
g_0 = np.pi + np.pi / 2 + phi_0
t = a['time_norm'].to_numpy()
a['gamma'] = g_0 + w * t

a
```

```python
r = 1.01 * 10 ** (-2)
L = 4.7 * 10 ** (-2)
l0 = 8.865 * 10 ** (-2)
j = 16.16 * 10 ** (-2)
d0 = 2.17 * 10 ** (-2)

g = a['gamma'].to_numpy()
xc = r * np.cos(g)
yc = r * np.sin(g)
x = xc - np.sqrt(L ** 2 - yc ** 2)
a['x'] = x
a['lenth'] = j - np.abs(a['x'].to_numpy())
a['V'] = d0 * a['lenth']
a
```

```python
plt.plot(a['time'], a['V'])
plt.grid()
plt.figure(figsize=(12, 9))
plt.show()
```
