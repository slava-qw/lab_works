```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.ticker as ticker

# разомкнутый конец
df1 = pd.read_csv('1.csv', sep='\t')
df1
```

```python
#1.1
fig, ax = plt.subplots()

ax.set(xlabel='f (Hz)', ylabel='|Z| (Ohm)')
ax.grid()
ax.scatter(df1['f (Hz)'], df1['|Z| (Ohm)'], s=10)
```

```python
#1.2
fig, ax = plt.subplots()

ax.set(xlabel='f (Hz)', ylabel='Arg Z (Ohm)')
ax.grid()
ax.scatter(df1['f (Hz)'], df1['Arg Z (Ohm)'], s=10)
```

```python
#1.3
fig, ax = plt.subplots()

ax.set(xlabel='f (Hz)', ylabel='Re Z (Ohm)')
ax.grid()
ax.scatter(df1['f (Hz)'], df1['Re Z (Ohm)'], s=10)
```

```python
#1.4
fig, ax = plt.subplots()

ax.set(xlabel='f (Hz)', ylabel='Im Z (deg)')
ax.grid()
ax.scatter(df1['f (Hz)'], df1['Im Z (deg)'], s=10)
```
