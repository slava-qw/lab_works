```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/physics/терма лабы/seriy', sep='\t')
df.tail()
```

```python
df = df[df.columns[[0, 1, 14, 5, 4, 11, 8, 2, 13, 9, 10, 7, 3, 12, 6]]]
df
```

```python
df.columns = ['time', 'temp_1', 'temp_2', 'temp_3', 'temp_4', 'temp_5', 'temp_6', 'temp_7', 'temp_8', 'temp_9', 'temp_10', 'temp_11', 'temp_12', 'temp_13', 'temp_14']
df
```

```python
df['temp_1'].describe()
```

```python
idx = df['temp_1'].argmax()
idx
```

```python
series_of_max_temp = df.loc[idx]
series_of_max_temp
```

```python
df_mt = series_of_max_temp.to_frame()
df_mt
```

```python
plt.figure(figsize=(10, 8))
plt.xlabel(r'$t, s$')
plt.ylabel(r'$temp, {}^{\circ}C$')
colors = ['crimson', 'red', '#bc13fe', 'orange', '#fdaa48', 'yellow', '#b0dd16', 'green', '#047495', '#8e82fe', 'blue', '#069af3', '#380282', '#4b006e']
for i in range(1, 15):
    plt.plot(df['time'], df[f'temp_{i}'], label=f'temp_{i}', color=colors[i - 1])

plt.legend()
plt.grid()
plt.scatter([657.47] * 100, np.linspace(27.5, 46, 100))
plt.savefig('temp(time).png')
```

```python
x = np.asarray([0.01 * i for i in range(0, 14)])
temp = np.asarray(series_of_max_temp[1:])

plt.figure(figsize=(10, 8))
plt.xlabel(r'$x$, м')
plt.ylabel(r'$temp, K$')


smooth_x = np.linspace(0, 0.13, 100)
t_0, a = 304.274, 113.954  # from Desmos for lin approx
t_0_sh, a_sh, s, x_0 = 304.736, 11.7076, 8.21433, 0  # from Desmos for sinh approx

smooth_temp = t_0 + a * smooth_x
sh_temp = t_0_sh + a_sh * np.sinh(s * (smooth_x - x_0))


plt.plot(x[::-1], 273 + temp, label=r'$t(x)$')
# plt.plot(smooth_x, smooth_temp, label=f'lineral approximation with $a$ = {a}, $t_0$ = {t_0}')
plt.plot(smooth_x, sh_temp, label=f'non-lineral approximation with $a$ = {a_sh}, $t_0$ = {t_0_sh} \n x_0 = {x_0}, psi = {s}')
plt.grid()
plt.legend()
```

```python
lin_temp = pd.DataFrame(273 + temp, x[::-1])
lin_temp
```
